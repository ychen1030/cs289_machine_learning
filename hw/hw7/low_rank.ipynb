```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

%matplotlib inline 
```

# Part a)

```python
def low_rank_approximation(X, rank):
    # YOUR CODE GOES HERE
    u, s, vt = np.linalg.svd(X)
    s_head = np.zeros((len(u), len(vt)))
    s_head[np.arange(rank), np.arange(rank)] = s[:rank]
    return np.matmul(np.matmul(u, s_head), vt)
```

```python
face = imread("./data/face.jpg")
low_rank_5 = low_rank_approximation(face, 5)
low_rank_20 = low_rank_approximation(face, 20)
low_rank_100 = low_rank_approximation(face, 100)

f, (ax1, ax2, ax3, ax4) = plt.subplots(nrows= 1, ncols= 4, figsize = (20, 5)) 
ax1.imshow(face, cmap='gray')
ax1.set_title("original")
ax2.imshow(low_rank_5, cmap='gray')
ax2.set_title("rank 5")
ax3.imshow(low_rank_20, cmap='gray')
ax3.set_title("rank 20")
ax4.imshow(low_rank_100, cmap='gray')
ax4.set_title("rank 100")
```

# Part b)

```python
sky = imread("./data/sky.jpg")
low_rank_5 = low_rank_approximation(sky, 5)
low_rank_20 = low_rank_approximation(sky, 20)
low_rank_100 = low_rank_approximation(sky, 100)

f, (ax1, ax2, ax3, ax4) = plt.subplots(nrows= 1, ncols= 4, figsize = (20, 5)) 
ax1.imshow(sky, cmap='gray')
ax1.set_title("original")
ax2.imshow(low_rank_5, cmap='gray')
ax2.set_title("rank 5")
ax3.imshow(low_rank_20, cmap='gray')
ax3.set_title("rank 20")
ax4.imshow(low_rank_100, cmap='gray')
ax4.set_title("rank 100")
```

# Part c)

```python
def mse(img1, img2):
    # YOUR CODE GOES HERE 
    return np.sum((img1 - img2)**2)
```

```python
mse_face, mse_sky = [], []
for i in range(1, 101):
    mse_face.append(mse(face, low_rank_approximation(face, i)))
    mse_sky.append(mse(sky, low_rank_approximation(sky, i)))

plt.plot(range(1, 101), mse_face, label = 'face')
plt.plot(range(1, 101), mse_sky, label = 'sky')
plt.xlabel("rank")
plt.ylabel("mse")
plt.legend()
plt.show()
```
